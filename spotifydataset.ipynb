```python
#multiple feature plots based on target
```

```python
data.columns
```

```python
interest = ['acousticness', 'danceability', 'duration_ms', 'energy',
       'instrumentalness', 'key', 'liveness', 'loudness', 'mode',
       'speechiness', 'tempo', 'time_signature', 'valence','target']
```

```python
for  names in interest:
  pos_data = data[data["target"] == 1][names]
  neg_data = data[data["target"] == 0][names]
  plt.figure(figsize=(12,7))
  sns.histplot(pos_data, bins=30, label="positive", color="green")
  sns.histplot(neg_data, bins=30, label="negative", color="red")
  plt.legend()
  plt.title(f"positive and negative histogram plots for {names} ")
```
